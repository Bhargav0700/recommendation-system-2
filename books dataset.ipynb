```python
import pandas as pd
import numpy as np
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
books= pd.read_csv('book.csv',encoding='latin-1')
```

```python
books
```

```python
books.rename( columns={'Unnamed: 0':'Index'}, inplace=True )
```

```python
books
```

```python
books.duplicated().sum()
```

```python
books.sort_values('User.ID')
```

```python
books = books.rename(columns={'User.ID': 'userid','Book.Title':'Title','Book.Rating':'Rating'})
```

```python
books
```

```python
books1=books.pivot_table(index='userid',
          columns='Title',
    values='Rating')
```

```python
books1
```

```python
books1.fillna(0,inplace=True)
```

```python
books1
```

```python
books2 = 1 - pairwise_distances(books1.values, metric='cosine')
```

```python
books2
```

```python
books2=pd.DataFrame(books2)
```

```python
books2
```

```python
books2.index = books.userid.unique()
books2.columns = books.userid.unique()
```
